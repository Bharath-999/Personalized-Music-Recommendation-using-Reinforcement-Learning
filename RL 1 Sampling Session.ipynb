```python
import glob
import pandas as pd
import numpy as np
```

```python
data_file_list = glob.glob("Data/training_set/*")
session_data_df = pd.read_csv(data_file_list[1])
session_data_df.head()
```

```python
list(session_data_df.columns)
```

```python
def sample_data(filename, sample_size):
    df = pd.read_csv(filename)
    twenty_session_length_session_ids = df[df['session_length'] == 20].session_id.unique()
    random_sampled_session_ids = np.random.choice(twenty_session_length_session_ids, sample_size, False)
    sampled_df = df[df.session_id.isin(random_sampled_session_ids)]
    return sampled_df
```

```python
from tqdm import tqdm
sampled_batch_df = pd.DataFrame(columns = session_data_df.columns)
sample_size = int(10000/len(data_file_list))
for file_name in tqdm(data_file_list):
    sampled_batch_df = sampled_batch_df.append(sample_data(file_name, sample_size))
sampled_batch_df.head()
```
